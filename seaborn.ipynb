## seaborn tutorial


```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.corr()
#corr=correlation
```

```python
sns.heatmap(df.corr())
```

## univariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
#hex=hexagon
# x and y are showing histograms
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#reg=regression
```

```python
#pairplot-helps to create scatter plots
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
#distplot-helps to create histograms
#distplot helps us to check distribution of the columns feature
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
#bar plot
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(y='sex',x='total_bill',data=df)
```

```python
##boxplot
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=df,palette='rainbow')
```

```python
#categorize my data based on some other categories
sns.boxplot(x='day',y='total_bill',hue='sex',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
#violin plot
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='rainbow',kde=False)
```
